We're going to scope back the question. Given that we have 120, non-oathed meta wolfpack ships, are there any nearby values of +RLD that will grant us an extra torpedo shot?

```python
ssv_stats.iloc[[20, 5, 3]]
```

The wiki states that submerged time is

$$ T = \frac{\rm Oxy}{10}. $$

```python
def g7e_reload_modifier(RLD, RLD_BONUS):
    assert RLD_BONUS < 2
    G7E_TORPEDO_RELOAD = 13.40
    return G7E_TORPEDO_RELOAD * np.sqrt(200 / (RLD * (1 + RLD_BONUS) + 100))
```

### U-96

```python
U96_OXY = 188
U96_RLD = 109
OXY = [U96_OXY, U96_OXY + OXY_AUX]
# Non Oathed
RLD = [U96_RLD, 115]

# With two RLD meowfficers (14 * 2). 40 is if two meowfficers have all three reload traits.
RLD_MEOW = [0, 14, 28, 40]
RLD_BONUS = .21
# ACTIVE_TIME = OXY / 10 + 5

for oxy in OXY:
    for rld in RLD:
        for meow in RLD_MEOW:
            print("OXY: ", oxy, ", RLD: ", rld, ", RLD_MEOW: ", meow)
            ACTIVE_TIME = oxy / 10 + 5
            salvos = ACTIVE_TIME / g7e_reload_modifier(rld + meow, RLD_BONUS)
            print(salvos)
```

TL;DR; Because U-96 has a self reload buff, she can fire 2 salvos without requiring the extra oxygen from a Snorkel, if she has enough rld buffs from oath/meowfficers. It does not look like she can realistically hit 3 salvos.
