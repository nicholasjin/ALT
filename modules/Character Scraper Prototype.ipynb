```python
from bs4 import BeautifulSoup
import requests
```

```python
target_website = "https://azurlane.koumakan.jp/Akashi"
result = requests.get(target_website)
assert result.status_code == 200
```

```python
class Character:
    def __init__(self, data):
        self.stats = data
```

```python
def save_char(url):
    result = requests.get(url)
    assert result.status_code == 200
    soup = BeautifulSoup(result.content)
    print(soup.prettify())
```
