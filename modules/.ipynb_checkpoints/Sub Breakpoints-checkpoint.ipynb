# Meta Wolfpack (U96, U47, U81)

U96:
- Increases this boat's Torpedo Crit Rate by 10% (20%). At the start of the battle: fires a torpedo barrage (damage and pattern of barrage is based on skill level). 
- Increases Accuracy by 1% (10%) and Crit DMG by 5% (15%) for all your Iron Blood SSs. Additionally, for each Iron Blood SS in your fleet: increases this boat's Accuracy, RLD, and EVA by 1.5% (6%). 
- When entering the battle, triggers Full Barrage - U-96 I (II) 

U47:
- When this boat hits an enemy with its Torpedoes: for 5s,the enemy hit will take 20% (40%) more DMG from this boat's Torpedoes. 
- When entering the map, if U-47 is the sole unit of the submarine fleet: increases damage dealt to Battleships (Battlecruisers and Aviation Battleships are not counted) by 5% (20%), increases Search Level by 1 and increases Ammo by 2.
- When entering the battle, triggers Full Barrage - U-47 I （II)

U-81:
- Increase damage dealt to CVs by 5% (20%). When own torpedo hits an enemy ship: 5% (15%) chance to cause Flooding for 24 second, inflicting damage every 3 second (Damage depends on torpedo type/attribute). If effect is applied on an enemy ship that already has Flooding, refreshes Flooding duration. 
- Increases own Evasion by 20% (40%). During battle, increases own Torpedo stat, Reload and Accuracy by 2.5% (7%) per Iron Blood submarine in the same fleet.
